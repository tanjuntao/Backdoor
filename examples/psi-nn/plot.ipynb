```python
def plot_acc(data: dict):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    x_epochs = list(range(50))

    for key, values in data.items():
        ax.plot(x_epochs, values, label=key)
    
    ax.grid(True, linestyle="-.")
    ax.set_ylabel("Acc", labelpad=5, loc="center")
    ax.set_xlabel("Epoch", labelpad=5, loc="center")
    plt.legend(loc="best")
    # plt.savefig(f"embedding_norm.pdf")
    plt.show()
    plt.close()
```
