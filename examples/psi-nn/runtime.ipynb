```python
def layer_count(history, layer_dict):
    for epoch_layers in history:
        for layer in epoch_layers:
            if layer not in layer_dict:
                layer_dict[layer] = 1
            else:
                layer_dict[layer] += 1
    
    sorted_layer_dict = {
            k: v
            for k, v in sorted(
                layer_dict.items(),
                key=lambda item: item[1],
                reverse=True,
            )
        }
    return sorted_layer_dict
```

```python
def size(data, model):
    sums = 0
    for val in data:
        sums += len(val)
    if model == "resnet18":
        return sums / (50 * 18)
    elif model == "vgg13":
        return sums / (50 * 11)    
    elif model == "lenet":
        return sums / (50 * 5)
    else:
        return sums / (50 * 3)
```
