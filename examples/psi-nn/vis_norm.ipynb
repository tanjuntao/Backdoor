```python
import matplotlib.pyplot as plt
import pickle
```

```python
# model_dir = "/home/1002tjt/models/cifar10_resnet18"
model_dir = "/home/1002tjt/models/cifar10_resnet18_concat"

with open(f"{model_dir}/history_bottom_ouput_norms.pkl", "rb") as f:
    history_bottom_ouput_norms = pickle.load(f)
    print(history_bottom_ouput_norms)
    print(len(history_bottom_ouput_norms))

with open(f"{model_dir}//history_embedding_norms.pkl", "rb") as f:
    history_embedding_norms = pickle.load(f)
    print(history_embedding_norms)
    print(len(history_embedding_norms))

with open(f"{model_dir}/history_embedding_grad_norms.pkl", "rb") as f:
    history_embedding_grad_norms = pickle.load(f)
    print(history_embedding_grad_norms)
    print(len(history_embedding_grad_norms))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x_epochs = list(range(len(history_bottom_ouput_norms)))


for label in history_bottom_ouput_norms[0].keys():
    target_label_norms = [epoch_norms[label] for epoch_norms in history_bottom_ouput_norms]
    ax.plot(x_epochs, target_label_norms, label=label)


ax.grid(True, linestyle="-.")
# ax.set_title("Layer gradient")
ax.set_ylabel("Bottom output norm", labelpad=5, loc="center")
ax.set_xlabel("Epoch", labelpad=5, loc="center")
plt.legend(loc="best")
# plt.savefig(f"embedding_norm.pdf")
plt.show()
plt.close()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x_epochs = list(range(len(history_embedding_norms)))


for label in history_embedding_norms[0].keys():
    target_label_norms = [epoch_norms[label] for epoch_norms in history_embedding_norms]
    ax.plot(x_epochs, target_label_norms, label=label)


ax.grid(True, linestyle="-.")
# ax.set_title("Layer gradient")
ax.set_ylabel("Embedding norm", labelpad=5, loc="center")
ax.set_xlabel("Epoch", labelpad=5, loc="center")
plt.legend(loc="best")
# plt.savefig(f"embedding_norm.pdf")
plt.show()
plt.close()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x_epochs = list(range(len(history_embedding_norms)))

for label in history_embedding_grad_norms[0].keys():
    target_label_norms = [epoch_norms[label] for epoch_norms in history_embedding_grad_norms]
    ax.plot(x_epochs, target_label_norms, label=label)


ax.grid(True, linestyle="-.")
# ax.set_title("Layer gradient")
ax.set_ylabel("Embedding grad norm", labelpad=5, loc="center")
ax.set_xlabel("Epoch", labelpad=5, loc="center")
plt.legend(loc="best")
# plt.savefig(f"embedding_norm.pdf")
plt.show()
plt.close()
```
